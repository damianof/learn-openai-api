```python
# Example: Predicting House Prices
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

areas = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
prices = np.array([250, 450, 550, 680, 770, 890, 980, 1150, 1280, 1400])

plt.scatter(areas, prices)
plt.xlabel('Area (sq. ft)')
plt.ylabel('Price ($)')
plt.title('House Prices')
plt.show()

model = LinearRegression()
model.fit(areas.reshape(-1, 1), prices)

new_area = np.array([[350]])
predicted_price = model.predict(new_area)
print(f"Predicted Price for a 350 sq. ft house: ${predicted_price[0]:.2f}")

# Generate a range of areas for plotting the line
plot_areas = np.linspace(0, 1200, 100)
# Predict the prices for the plot areas
predicted_prices = model.predict(plot_areas.reshape(-1, 1))

plt.scatter(areas, prices)
plt.plot(plot_areas, predicted_prices, color='red', label='Prediction')
plt.xlabel('Area (sq. ft)')
plt.ylabel('Price ($)')
plt.title('House Prices with Prediction')
plt.legend()
plt.show()
```
