```python
import matplotlib.pyplot as plt

# Sample data
x_values = [0, 1, 2, 3, 4, 5]
y_values = [0, 1, 4, 9, 16, 25]

plt.plot(x_values, y_values)

# Adding title and labels for the axes
plt.title('A Simple Line Plot')
plt.xlabel('X values')
plt.ylabel('Y values')

# Display the plot
plt.show()
```
